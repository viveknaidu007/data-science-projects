Chess (King-Rook v/s King)

chess_image


This dataset is a well-known dataset used in the field of artificial intelligence and machine learning for studying chess endgames. It is specifically focused on the endgame scenario where there are only three pieces left on the board: a king, a rook, and another king.


details


The dataset consists of various positions and their corresponding outcomes. Each position is described by the positions of the three pieces on the chessboard. The positions are represented using algebraic notation, where each square on the board is identified by a letter (from "a" to "h") for the file and a number (from 1 to 8) for the rank. For example, the square in the bottom left corner is a1, and the square in the top right corner is h8.
